 Task 1: Predictive Modeling

Build a regression model to predict the
aggregate rating of a restaurant based on
available features

```python
df.columns
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
#Feature and target
x = df.drop(['Aggregate rating', 'Rating color', 'Rating text', 'Restaurant Name', 'restaurant_name_length'],axis=1)
y = df['Aggregate rating']
```

```python
non_numeric_cols = x.select_dtypes(include=['object']).columns
print(f"Non-numeric columns: {non_numeric_cols}")
```

```python
print("Unique values in 'Is delivering now':", x['Is delivering now'].unique())
print("Unique values in 'Switch to order menu':", x['Switch to order menu'].unique())
```

```python
le = LabelEncoder()
```

```python
x['Is delivering now'] = le.fit_transform(x['Is delivering now'])
x['Switch to order menu'] = le.fit_transform(x['Switch to order menu'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
random_Model = RandomForestRegressor(random_state=42)
random_Model.fit(x_train, y_train)
```

```python
y_pred = random_Model.predict(x_test)
```

```python
 R2 = r2_score(y_pred, y_test)
```

```python
mse = mean_squared_error(y_pred, y_test) 
```

```python
print(f" R2 score: {R2}")
```

```python
print(f" Mean Squared error: {mse}")
```

Conclusion

The regression model demonstrates excellent performance in predicting the aggregate ratings of restaurants based on the available features. 

RÂ² Score: The RÂ² value of 0.96 indicates that the model explains 96% of the variance in the aggregate ratings. This reflects a strong relationship between the features and the target variable.

Mean Squared Error (MSE): The MSE value of 0.089 suggests that the average squared difference between the predicted and actual values is minimal, signifying that the predictions are highly accurate.

Model Effectiveness: The combination of high RÂ² and low MSE highlights that the model has effectively captured the underlying patterns in the data and is suitable for this regression task.

Feature Contribution: The preprocessing steps, including encoding categorical variables and removing irrelevant or low-variance columns, contributed significantly to the model's success by ensuring high-quality input data.
